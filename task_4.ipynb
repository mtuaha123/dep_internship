```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler , LabelEncoder
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df=pd.read_csv("/content/CloudWatch_Traffic_Web_Attack.csv")
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
ctgr_col=['src_ip','src_ip_country_code','protocol','dst_ip','rule_names','observation_name','source.meta','source.name','detection_types']
for i in ctgr_col:
  encoder = LabelEncoder()
  df[i]=encoder.fit_transform(df[i])
```

```python
df
```

```python
x=df.drop(columns=['creation_time','end_time','time'])
```

```python
x['src_ip_country_code'].plot(kind='hist',bins=30)
plt.title('country_code')
plt.xlabel('value')
plt.ylabel('freq')
plt.show()
```

```python
scaler=StandardScaler()
xscaled=scaler.fit_transform(x)
```

```python
xscaled
```

```python
x_train,x_test=train_test_split(xscaled,test_size=0.2,random_state=33)
```

```python
model=IsolationForest(contamination=0.1,random_state=32)
model.fit(x_train)
```

```python
y_pred=model.fit_predict(x_test)
y_pred
```

```python
y_pred=[ 1 if x==1 else 0 for x in y_pred]
y_pred
```

```python
y_test=[1]*len(y_pred)
```

```python
print("clasification report:")
print(classification_report(y_test,y_pred))
```
